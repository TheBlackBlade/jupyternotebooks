```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv/master.csv', delimiter=';')
df2 = df[df.wr!='total']
df2 = df2[df.sim_tool!='None']
df2['sum_rel'] = df2.summe/df2.installed_p
```

```python
g = sns.boxplot(x='module',y='sum_rel', data=df2, hue='sim_tool')
g = sns.swarmplot(x='module', y='sum_rel', data=df2, hue='sim_tool', dodge=True, palette='Set1')
plt.show()
```

```python
g = sns.violinplot(x='module',y='sum_rel', data=df2, hue='sim_tool', scale='width')
g = sns.swarmplot(x='module', y='sum_rel', data=df2, hue='sim_tool', dodge=True, palette='Set1')
plt.show()
```

```python
g = sns.pairplot(data=df2, kind='reg', hue='sim_tool', vars=['summe', "sum_rel", 'installed_p', 'jan', 'feb', 'aug'], palette='Set1')
plt.show()
```

```python
box_ax1 = df2.boxplot('sum_rel','wr',rot = 30,figsize=(5,6))
plt.savefig('sum_rel_for_wr.png')
box_ax2 = df2.boxplot('sum_rel','inv',rot = 30,figsize=(5,6))
plt.savefig('sum_rel_for_inv.png')
box_ax3 = df2.boxplot('sum_rel','module',rot = 30,figsize=(5,6))
plt.savefig('sum_rel_for_module.png')
```
